```python
import yfinance as yf
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python

# List of stock symbols for the given companies
stocks = [
    'BAJFINANCE.NS',    # Bajaj Finance
    'HDFCBANK.NS',      # HDFC Bank
    'HINDUNILVR.NS',    # Hindustan Unilever
    'ICICIBANK.NS',     # ICICI Bank
    'INFY.NS',          # Infosys
    'KOTAKBANK.NS',     # Kotak Mahindra Bank
    'LT.NS',            # Larsen & Toubro
    'RELIANCE.NS',      # Reliance Industries
    'TCS.NS',           # Tata Consultancy Services (TCS)
    'SBIN.NS'           # State Bank of India (SBI)
]

# Create an empty dictionary to store stock data
stock_data = {}

# Fetch data for each stock for the last 60 days
for stock in stocks:
    print(f"Fetching data for {stock}...")
    data = yf.download(stock, period="60d")  # Download 60 days of data
    stock_data[stock] = data

    # Optionally, save each stock's data to a CSV file
    data.to_csv(f"{stock}_last_60_days.csv")

```

```python
BF = pd.read_csv("/content/drive/MyDrive/stocks/BAJFINANCE.NS_last_60_days.csv")
HF = pd.read_csv("/content/drive/MyDrive/stocks/HDFCBANK.NS_last_60_days.csv")
HUL = pd.read_csv("/content/drive/MyDrive/stocks/HINDUNILVR.NS_last_60_days.csv")
IB = pd.read_csv("/content/drive/MyDrive/stocks/ICICIBANK.NS_last_60_days.csv")
INF = pd.read_csv("/content/drive/MyDrive/stocks/INFY.NS_last_60_days.csv")
KMB = pd.read_csv("/content/drive/MyDrive/stocks/KOTAKBANK.NS_last_60_days.csv")
LT = pd.read_csv("/content/drive/MyDrive/stocks/LT.NS_last_60_days.csv")
RI = pd.read_csv("/content/drive/MyDrive/stocks/RELIANCE.NS_last_60_days.csv")
SBI = pd.read_csv("/content/drive/MyDrive/stocks/SBIN.NS_last_60_days.csv")
TCS = pd.read_csv("/content/drive/MyDrive/stocks/TCS.NS_last_60_days.csv")
```

```python
ds={
    "BAJFINANCE.NS":BF,
    "HDFCBANK.NS":HF,
    "HINDUNILVR.NS":HUL,
    "ICICIBANK.NS":IB,
    "INFY.NS":INF,
    "KOTAKBANK.NS":KMB,
    "LT.NS":LT,
    "RELIANCE.NS":RI,
    "SBIN.NS":SBI,
    "TCS.NS":TCS
}
```

```python
for name,data in ds.items():
  data["Stock symbol"] = name
  data["Tomorrow"]=data["Close"].shift(-1)
  data["Target"] = (data["Close"] < data["Tomorrow"]).astype(int)

```

```python
dataset=pd.concat([BF,HF,HUL,IB,INF,KMB,LT,RI,SBI,TCS])
dataset
```

```python
dataset=dataset.rename(columns={"Price":"Date"})
dataset
```
